```python
import pandas as pd
import numpy as np
import csv
import math
from tqdm import tqdm
```

```python
df = pd.read_json("runs/06 Caleb Preprocessed/dataworks_sae_aave_validation_annotations.json")
df.head()
```

```python
df.shape
```

```python
counter = 0
with open('runs/06 Caleb Preprocessed/data.csv', 'a', encoding="utf-8") as f:
    writer = csv.writer(f)
    for head in tqdm(df):
        accept_list = df[head]['accept']
        accept_list = [x for x in accept_list if x != None]
        if np.sum(accept_list) == len(accept_list) or df[head]['aave_human']:
            if np.sum(accept_list) == len(accept_list):
                writer.writerow([head, df[head]['aave_synthetic']])
            else:
                writer.writerow([head, df[head]['aave_human'][0]])
```
